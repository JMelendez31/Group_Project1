```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
life_expectancy = pd.read_csv('Life-Expectancy-Data.csv')
life_expectancy
```

```python
size_bins = [0, 5000, 10000, 25000, 50000, 75000, 115000]
category_labels = ["<$5,000", "$5,000-$10,000", "$10,000-$25,000","$25,000-$50,000","$50,000-$75,000","$75,000-$115,000"]

life_expectancy["GDP Ranges"] = pd.cut(life_expectancy["GDP_per_capita"],
                                                             size_bins, labels = category_labels)
life_expectancy
```

```python
gdp_ranges_df=life_expectancy.groupby(["GDP Ranges", "Economy_status_Developed"])["Life_expectancy"].mean().reset_index()
```

```python
plt.bar(gdp_ranges_df["GDP Ranges"], gdp_ranges_df["Life_expectancy"])
plt.xticks(rotation = 90)
plt.show()
```
